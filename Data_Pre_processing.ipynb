```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
```

```python
dog_path = "/content/drive/Shareddrives/CIS 5500 Drive/AllDogs.csv"
dog_df = pd.read_csv(dog_path)
dog_df.head()
```

```python
# Dogs
df = dog_df.copy()
df = df.rename(columns={'id': 'dog_id'})
Dogs = df[['dog_id', 'org_id', 'name', 'age', 'sex', 'size']]

# change date format to 'YYYY-MM-DD'
# Dogs['posted'] = pd.to_datetime(Dogs['posted'], errors='coerce').dt.date

Dogs.head()
```

```python
# DogBreeds
df = dog_df.copy()
df = df.rename(columns={'id': 'dog_id'})
DogBreeds = df[['dog_id', 'breed_primary', 'breed_mixed', 'breed_unknown']]
DogBreeds.head()
```

```python
# DogAttributes
df = dog_df.copy()
df = df.rename(columns={'id': 'dog_id'})
DogAttributes = df[['dog_id', 'color_primary', 'color_secondary', 'coat', 'fixed', 'house_trained', 'special_needs', 'shots_current', 'env_children']].copy()

DogAttributes['color_primary'] = DogAttributes['color_primary'].str.split('/').str[0].str.strip()
DogAttributes['color_secondary'] = DogAttributes['color_secondary'].str.split('/').str[0].str.strip()

DogAttributes
```

```python
# import 2017_census
census_path = "/content/drive/Shareddrives/CIS 5500 Drive/2017_census.csv"
census_df = pd.read_csv(census_path)
census_df.head()
```

```python
# Tracts
df = census_df.copy()
df = df.rename(columns={'TractId': 'tract_id', 'State': 'state', 'County': 'county'})
Tracts = df[['tract_id', 'state', 'county']]
Tracts.head()
```

```python
# Demographics
df = census_df.copy()
df = df.rename(columns={
    'TractId': 'tract_id',
    'TotalPop': 'total_pop',
    'Men': 'men',
    'Women': 'women',
    'Hispanic': 'hispanic',
    'White': 'white',
    'Black': 'black',
    'Asian': 'asian',
    'Pacific': 'pacific'
})
Demographics = df[['tract_id', 'total_pop', 'men', 'women', 'hispanic', 'white', 'black', 'asian', 'pacific']]
Demographics.head()
```

```python
# EconomicsHousing
df = census_df.copy()
df = df.rename(columns={
    'TractId': 'tract_id',
    'Income': 'income',
    'IncomePerCap': 'income_per_cap',
    'Poverty': 'poverty',
    'ChildPoverty': 'child_poverty',
    'Employed': 'employed',
    'Unemployment': 'unemployment',
    'MeanCommute': 'mean_commute'
})
EconomicsHousing = df[['tract_id', 'income', 'income_per_cap',
                       'poverty', 'child_poverty',
                       'employed', 'unemployment', 'mean_commute']]
EconomicsHousing.head()
```

```python
# Shelters
df = dog_df.copy()
Shelters = df[['org_id', 'contact_city', 'contact_state', 'contact_zip']].copy()
Shelters = Shelters.drop_duplicates(subset=['org_id'])
Shelters = Shelters.rename(columns={'contact_city': 'city', 'contact_state': 'state', 'contact_zip': 'zip'})
Shelters = Shelters[['org_id', 'city', 'state', 'zip']]

# clean zip: turn NaN into empty string, strip spaces
Shelters['zip'] = Shelters['zip'].fillna('').astype(str).str.strip()

# Fix some misinputs
Shelters.loc[Shelters['org_id'] == 'WA581', ['city', 'state', 'zip']] = ['Seattle', 'WA', '98106']
Shelters.loc[Shelters['org_id'] == 'NM133', ['city', 'state', 'zip']] = ['Albuquerque', 'NM', '87108']

# Keep only:
#   - blank zips (still unknown)
#   - or 4â€“5 digit zips (US-style)
mask_valid = Shelters['zip'].eq('') | Shelters['zip'].str.fullmatch(r'\d{4,5}')
Shelters = Shelters[mask_valid].copy()

Shelters
```

```python
# export files to Google Drive
base = "/content/drive/Shareddrives/CIS 5500 Drive/"

Dogs.to_csv(base + "Dogs.csv", index=False, quoting=1)
DogBreeds.to_csv(base + "DogBreeds.csv", index=False, quoting=1)
DogAttributes.to_csv(base + "DogAttributes.csv", index=False, quoting=1)
Tracts.to_csv(base + "Tracts.csv", index=False, quoting=1)
Demographics.to_csv(base + "Demographics.csv", index=False, quoting=1)
EconomicsHousing.to_csv(base + "EconomicsHousing.csv", index=False, quoting=1)
Shelters.to_csv(base + "Shelters.csv", index=False, quoting=1)

```
